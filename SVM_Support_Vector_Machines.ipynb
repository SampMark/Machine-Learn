# **Modelando Máquinas de Vetores de Suporte (SVM)**

O modelo de *Support Vector Machines* (SVM) ou Máquina de Vetores de Suporte (SVM) é um algoritmo importante no arsenal do aprendizado de máquina, se destacam por sua capacidade de lidar com dados complexos e realizar classificações precisas.

O SVM tem ótimo desempenho em problemas de classificação, especialmente em conjuntos de dados não lineares de pequeno a médio porte. Ele funciona encontrando um hiperplano que separa os dados em diferentes categorias, mesmo quando os dados não são linearmente separáveis.

## **Como funciona o SVM?**

1. **Mapeamento de Dados para Espaços de Alta Dimensão**

Em muitos casos, os dados originais não podem ser separados linearmente em classes distintas. Para lidar com essa limitação, o SVM transforma os dados para um espaço de alta dimensão (espaço de características) onde eles possam ser mais facilmente separados. Por exemplo, imagine pontos distribuídos em uma linha sem separação clara entre classes. Ao projetar esses pontos em um espaço bidimensional, utilizando coordenadas como 𝑥 e 𝑥2, as classes podem se tornar linearmente separáveis. Essa transformação permite ao SVM trabalhar em problemas complexos onde os dados originalmente não apresentam separabilidade linear.

2. **Encontrando o Hiperplano Ótimo**

Após mapear os dados, o SVM busca o hiperplano que melhor separa as classes. Os dados devem ser transformados de forma que um separador possa ser desenhado como um hiperplano, em que os pontos de dados se enquadram em categorias diferentes, representando um conjunto de dados não separável linearmente.

Desta forma é possível transferir esses dados para um espaço de dimensão superior, por exemplo, mapeando-os para um espaço tridimensional. Após a transformação, o limite entre as duas categorias pode ser definido por um hiperplano. Num espaço tridimensional, o separador é mostrado como um plano. Este plano pode ser usado para classificar casos novos ou desconhecidos. Portanto, o algoritmo SVM produz um hiperplano ideal que categoriza novos exemplos.

Em outras palavras, se os dados estão dispostos em uma linha no espaço 1D e não são separáveis, podem ser projetados para um espaço 2D (como e ) onde se tornam separáveis. Após o mapeamento, o SVM  identifica o hiperplano ótimo que divide os dados em
classes distintas. Em um espaço 2D, esse hiperplano é uma linha, enquanto em um espaço 3D, é um plano.
Em dimensões mais altas, é chamado de hiperplano.

3. **Maximizando a Margem**

O SVM não busca apenas um hiperplano que separa as classes, mas sim aquele que maximiza a margem entre as duas categorias. A margem é definida como a distância entre o hiperplano e os pontos mais próximos de cada classe. Esses pontos são chamados de **vetores de suporte**, pois determinam a posição e a orientação do hiperplano.

Esse processo de maximização da margem confere ao SVM uma maior capacidade de generalização, minimizando erros ao classificar novos exemplos.

4.  **Kernelização**

Como nem sempre os dados podem ser separados linearmente, mesmo em espaços de alta dimensão. Para resolver isso, o SVM utiliza uma técnica chamada kernelização, que aplica funções matemáticas (kernels) para transformar os dados em um novo espaço onde a separação linear seja possível, são eles:

* **Linear**: Mantém os dados no espaço original.
* **Polinomial**: Projeta os dados em uma dimensão polinomialmente maior.
* **RBF (Radial Basis Function)**: Projeta os dados em um espaço infinito com base em distâncias relativas.
* **Sigmoide**: Projeta os dados de forma não linear, similar a redes neurais.

Em suma, a kernelização permite que o SVM seja aplicado em dados não linearmente separáveis, tornando-o extremamente flexível e eficaz em uma ampla variedade de problemas.

Brevemente, resumo suas aplicações e limitações, vantagens e desvantagens, entre outros aspectos.

**Aplicações:**

SVMs são eficazes em uma variedade de tarefas, incluindo:

* **Classificação:**  Desde a categorização de imagens até a detecção de spam, as SVM demonstram excelente performance em problemas de classificação, especialmente quando a separação entre classes não é linear.
* **Regressão:**  Embora menos comum, as SVM também podem ser aplicadas em problemas de regressão, prevendo valores contínuos.
* **Detecção de Novidades:**  SVM podem identificar padrões incomuns ou *outliers* em conjuntos de dados.

**Performance e Escalabilidade**

O SVM é particularmente eficaze em conjuntos de dados de pequeno a médio porte (centenas a milhares de instâncias) com relações não lineares entre as variáveis, mas seu desempenho pode ser afetado em conjuntos de dados muito grandes, devido à complexidade computacional. O SVM pode apresentar desafios de escalabilidade.  Nesses casos, algoritmos alternativos como árvores de decisão ou redes neurais podem ser mais eficientes.

**Vantagens:**

* Eficazes em espaços de alta dimensão.
* Robustas a *outliers*.
* Versáteis, com diferentes funções kernel para modelar relações complexas.

**Desvantagens:**

* **Computacionalmente caros para grandes conjuntos de dados**: a complexidade treinamento de um modelo SVM envolve a otimização de um problema quadrático, que se torna computacionalmente caro à medida que o tamanho do conjunto de dados aumenta. O tempo de treinamento pode crescer significativamente com o número de amostras, tornando os SVM poucos eficientes para conjuntos de dados massivos. Embora o tempo de predição seja geralmente menor que o tempo de treinamento, este pode ser afetado pelo tamanho do conjunto de dados e pelo número de vetores de suporte, consumindo muita memória e impactando o desempenho.
* **Escolha da função kernel e dos hiperparâmetros pode se mostrar complexa**: o modelo utiliza funções kernel para mapear os dados em um espaço de alta dimensão. Existem diversos tipos de kernels (linear, polinomial, RBF, sigmoide), cada um com características e parâmetros próprios.
  * A escolha do kernel ideal depende do problema e dos dados, o que pode exigir experimentação e conhecimento especializado. Além disso, o SVM possui outros hiperparâmetros, como o parâmetro de regularização (C) e a tolerância para o critério de parada.Encontrar a combinação ideal de hiperparâmetros para um bom desempenho do modelo pode ser desafiador e demandar técnicas de otimização e validação cruzada.
  * A interpretação do impacto de cada hiperparâmetro no modelo pode ser complexa, exigindo conhecimento teórico e prático sobre SVMs.
* Caso se pretenda uma modelagem preditiva, o SVM não fornece diretamente estimativas de probabilidade.

### **Divisão dos dados: função `train_test_split`**

A função `train_test_split` do módulo `model_selection` do scikit-learn será utilizada para dividir os arrays (matrizes) em subconjuntos de treinamento e teste de forma aleatória.

**Parâmetros:**
* `X`: matriz de características (features).
* `y`: vetor de rótulos ou variáveis alvo.
* `test_size=0.2`: Define que 20% dos dados serão utilizados para o conjunto de teste.
* `random_state=4`: garante a reprodutibilidade da divisão dos dados, permitindo que os mesmos subconjuntos sejam obtidos em execuções subsequentes.
* `stratify=y`: assegura que a divisão mantenha a mesma proporção de classes em ambos os conjuntos, o que é especialmente importante em problemas de classificação para evitar vieses.

```python
# Dividir os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=4, stratify=y
)

# Exibir as dimensões dos conjuntos resultantes
print('Conjunto de treinamento:', X_train.shape, y_train.shape)
print('Conjunto de teste:', X_test.shape, y_test.shape)
```

## **Modelagem em SVM com kernel Radial Basis Function (RBF)**

---
A modelagem em SVM envolve o seguinte processo abstrato:

1. **Transformação inicial**: os dados são projetados para um espaço de alta dimensão (manualmente ou via kernel).
2. **Definição do hiperplano ótimo**: um separador é encontrado para dividir as classes.
3. **Maximização da margem**: a margem entre os vetores de suporte é maximizada, garantindo uma separação ideal.
4. **Classificação**: o modelo final utiliza o hiperplano para prever a classe de novos exemplos.

Como explicitado acima, o algoritmo Support Vector Machine utiliza o conceito de "*kernelling*" para mapear dados em um espaço dimensional mais alto, permitindo a classificação eficiente mesmo quando os dados não são linearmente separáveis. Essa transformação é realizada por meio de funções kernel, cada uma com suas próprias características e desempenho.

Existem diversas funções kernel disponíveis, como:

* **Linear**: eficaz para dados linearmente separáveis.
* **Polinomial:** mapeia os dados em um espaço polinomial, útil para dados com relações não lineares.
* ***Radial Basis Function*** ou Função de Base Radial (RBF): uma função popular que mapeia os dados em um espaço de dimensão infinita, geralmente oferecendo bom desempenho em diversos cenários.
* **Sigmoide**: similar à função sigmoide utilizada em redes neurais.

A escolha da função kernel ideal depende do conjunto de dados e do problema em questão. Uma vez que não existe uma regra universal para determinar a melhor opção, a experimentação com diferentes funções e a comparação dos resultados são cruciais para encontrar a função mais adequada.

```python
# Acessar o classificador SVM
from sklearn import svm
# Padroniza os dados
from sklearn.preprocessing import StandardScaler
# Cria um pipeline para facilitar o processamento sequencial
from sklearn.pipeline import make_pipeline

# Supondo que X_train e y_train já estejam definidos

# Criar um pipeline que inclui a padronização dos dados e o classificador SVM com kernel RBF
clf = make_pipeline(
    StandardScaler(),  # Padroniza os dados para ter média 0 e variância 1
    svm.SVC(kernel='rbf', C=1.0, gamma='scale')  # Inicializa o classificador SVM
)

# Treinar o modelo com os dados de treinamento
clf.fit(X_train, y_train)
```

```python
# Realizar previsões no conjunto de teste
yhat = clf.predict(X_test)

# Exibir as primeiras 5 previsões
print("As primeiras 5 previsões são:", yhat[:5])
```

### ***Trade-off* entre Precision e Recall**

A curva Precision-Recall fornece uma visão mais detalhada do desempenho em relação à classificação binária, especialmente quando as classes estão desbalanceadas. O *trade-off* entre *Precision* e *Recall* reflete a necessidade de ajustar o desempenho do modelo às prioridades do problema em questão. O gráfico abaixo fornece algumas informações para se encontrar o equilíbrio adequado, considerando-se as implicações dos erros de classificação no domínio da aplicação.

Precision e Recall rabalham em oposição. Ao se maximizar a precisão, tende-se a sacrificar a abrangência (recall), pois o modelo torna-se mais conservador nas previsões. E vice-versa, ao se maximizar o recall, a precisão pode cair, pois o modelo tenta classificar mais amostras como positivas. O *threshold* determina o ponto de corte usado para classificar uma amostra como positiva ou negativa:
* **Limiar baixo** (ex.: próximo de 0, aumenta o Recall): mais amostras serão classificadas como positivas, incluindo muitas da classe verdadeira, o que também aumenta o número de Falsos Positivos, reduzindo a Precision.
* **Limiar alto** (ex.: próximo de 1, aumenta a Precision): apenas as amostras com alta confiança são classificadas como positivas, mas algumas amostras positivas podem ser perdidas (Falsos Negativos), reduzindo o Recall.



```python
from sklearn.metrics import precision_recall_curve

# Obter pontuações da função de decisão
y_scores = clf.decision_function(X_test)

# precision_recall_curve utiliza rótulos binários (0/1 or -1/1),
# Especificar pos_label para indicar qual classe é considerada "positiva" para o cálculo da curva.
precisions, recalls, thresholds = precision_recall_curve(y_test, y_scores, pos_label=4)

plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
threshold = 0.6  # Exemplo de valor limite
plt.vlines(threshold, 0, 1.0, "k", "dotted", label="threshold")

plt.xlabel("Threshold")
plt.ylabel("Precision/Recall")
plt.title("Precision-Recall Curve")
plt.legend()
plt.grid(True)
plt.show()
```

## **Visualização Kernel Linear versus RBF**

* **Linear**: uma linha reta separa as duas classes, para dados linearmente separáveis.
* **RBF**: separação não linear ajusta-se às curvas formadas pelos dados, adequado para dados mais complexos.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs

# 1. Gerar dados sintéticos
X, y = make_blobs(n_samples=100, centers=2, random_state=42, cluster_std=1.5)

# 2. Treinar o modelo SVM com kernel linear e RBF
svm_linear = SVC(kernel="linear", C=1)
svm_rbf = SVC(kernel="rbf", C=1, gamma=0.5)

svm_linear.fit(X, y)
svm_rbf.fit(X, y)

# 3. Função para visualização do plano de decisão
def plot_decision_boundary(svm, X, y, ax, title):
    # Criar grade de pontos para o gráfico
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))
    # Calcular valores da função de decisão
    Z = svm.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plotar os dados e o plano de decisão
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors="k", s=50)
    ax.contour(xx, yy, Z, levels=[-1, 0, 1], linestyles=["--", "-", "--"], colors="k")
    ax.scatter(svm.support_vectors_[:, 0], svm.support_vectors_[:, 1], s=200,
               facecolors="none", edgecolors="red", label="Vetores de Suporte")
    ax.set_title(title)
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")
    ax.legend()

# 4. Criar gráficos lado a lado
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Kernel Linear
plot_decision_boundary(svm_linear, X, y, axes[0], title="SVM com Kernel Linear")

# Kernel RBF
plot_decision_boundary(svm_rbf, X, y, axes[1], title="SVM com Kernel RBF")

plt.tight_layout()
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.datasets import load_iris

# 1. Carregar o dataset Iris
iris = load_iris()
X = iris["data"][:, (2, 3)]  # Usar comprimento e largura das pétalas
y = (iris["target"] == 2).astype(int)  # Iris virginica como classe positiva

# 2. Ajustar dois modelos SVM com diferentes valores de C
svm_clf_low_c = SVC(kernel="linear", C=1)
svm_clf_high_c = SVC(kernel="linear", C=100)
svm_clf_low_c.fit(X, y)
svm_clf_high_c.fit(X, y)

# 3. Função para calcular margens e suportes
def plot_svm_decision_boundary(svm_clf, X, y, ax, title):
    # Plotar os dados
    ax.scatter(X[:, 0][y == 1], X[:, 1][y == 1], color="green", s=60, marker="^", label="Iris virginica")
    ax.scatter(X[:, 0][y == 0], X[:, 1][y == 0], color="blue", s=60, marker="s", label="Iris versicolor")

    # Gerar grade para decisão
    x0s = np.linspace(X[:, 0].min(), X[:, 0].max(), 200)
    x1s = np.linspace(X[:, 1].min(), X[:, 1].max(), 200)
    x0, x1 = np.meshgrid(x0s, x1s)
    X_new = np.c_[x0.ravel(), x1.ravel()]
    y_pred = svm_clf.decision_function(X_new).reshape(x0.shape)

    # Plotar fronteiras de decisão
    ax.contour(x0, x1, y_pred, colors="black", levels=[-1, 0, 1], alpha=0.7, linestyles=["--", "-", "--"])

    # Destacar vetores de suporte
    ax.scatter(
        svm_clf.support_vectors_[:, 0],
        svm_clf.support_vectors_[:, 1],
        s=200, facecolors="none", edgecolors="gray", label="Vetores de Suporte"
    )
    ax.set_title(title)
    ax.set_xlabel("Petal length")
    ax.set_ylabel("Petal width")
    ax.legend()

# 4. Criar gráficos lado a lado
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

plot_svm_decision_boundary(svm_clf_low_c, X, y, axes[0], title="C = 1")
plot_svm_decision_boundary(svm_clf_high_c, X, y, axes[1], title="C = 100")

plt.tight_layout()
plt.show()
```

O gráfico acima compara claramente o impacto do parâmetro C, efeito da regularização em aprendizado de máquina, em um classificador SVM.

* **C baixo** (margens amplas): Maior flexibilidade, maior número de violações.
* **C alto** (margens estreitas): Classificação mais rígida, menor número de violações.
